```python
import warnings; warnings.simplefilter('ignore')
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, numpy as np
import matplotlib.cm as cm
from collections import Counter
plt.style.use('bmh'); sns.set()
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv', encoding='latin-1')
df
```

```python
# Menghapus index berupa "Person ID"
df.reset_index(drop=True, inplace=True)
df
```

```python
# Pre-processing berupa mengecek data duplikasi
print(df.shape)
df.duplicated().sum()
```

```python
# Hapus baris duplikat, simpan ke DataFrame baru
df2 = df.drop_duplicates()
df2
```

```python
#Deteksi banyaknya outlier pada variabel 'Age'
z_score =  (df2['Age'] - df2['Age'].mean())/df2['Age'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi banyaknya outlier pada variabel 'Sleep Duration'
z_score =  (df2['Sleep Duration'] - df2['Sleep Duration'].mean())/df2['Sleep Duration'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi banyaknya outlier pada variabel 'Quality of Sleep'
z_score =  (df2['Quality of Sleep'] - df2['Quality of Sleep'].mean())/df2['Quality of Sleep'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi banyaknya outlier pada variabel 'Physical Activity Level'
z_score =  (df2['Physical Activity Level'] - df2['Physical Activity Level'].mean())/df2['Physical Activity Level'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi banyaknya outlier pada variabel 'Stress Level'
z_score =  (df2['Stress Level'] - df2['Stress Level'].mean())/df2['Stress Level'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi banyaknya outlier pada variabel 'Heart Rate'
z_score =  (df2['Heart Rate'] - df2['Heart Rate'].mean())/df2['Heart Rate'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
# Menghilangkan outlier
df3 = df2[(z_score < 3) & (z_score > -3)].copy()
print(f"Sesudah menghilangkan outlier: {df3.shape}")
```

```python
# Memanggil kembali data yang sudah dibuang outliernya
df3
```

```python
#Deteksi banyaknya outlier pada variabel 'Daily Steps'
z_score =  (df3['Daily Steps'] - df3['Daily Steps'].mean())/df3['Daily Steps'].std()
print("banyaknya outlier = ",sum(abs(z_score) > 3))
```

```python
#Deteksi missing value
df3.isnull().sum()
```

```python
#Meghitung persentase missing
(df3.isnull().sum()/len(df3)).to_frame('persentase missing')
```

```python
#Mengganti data missing pada variabel Sleep Disorder
df4 = df3.fillna({'Sleep Disorder': df3['Sleep Disorder'].mode()[0]})
print(df3['Sleep Disorder'].mode()[0])
df4
```

```python
# Mengecek Korelasi
df_corr = df4[['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']].corr()
df_corr
```

```python
# Heatmap untuk menyelidiki korelasi
df_corr2 = df_corr
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr2,
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 14}, square=True);
```

```python
#Menyimpan data preprocessing
import pandas as pd
df_preprocessed = pd.DataFrame(df4)
```

```python
# Simpan ke file CSV di Google Colab
df_preprocessed.to_csv('preprocessed_data.csv', index=False)

print("File CSV berhasil disimpan!")
```
