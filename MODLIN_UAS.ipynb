```python
df_numeric = df.select_dtypes(include=np.number)

# Define the list of columns you want to analyze for outliers
columns_without_opt = ['Heart Rate'] # Assuming these are the columns you want to check

df_numeric_filtered = df_numeric[columns_without_opt]
Q1 = df_numeric[columns_without_opt].quantile(0.25)
Q2 = df_numeric[columns_without_opt].quantile(0.50)
Q3 = df_numeric[columns_without_opt].quantile(0.75)
IQR = Q3 - Q1
# Ensure alignment by using .loc and matching indices
Banyaknya_Outlier = ((df_numeric[columns_without_opt] <= (Q1 - 1.5 * IQR)) | (df_numeric[columns_without_opt] >= (Q3 + 1.5 * IQR))).sum()
print(Banyaknya_Outlier)
```

```python
df_numeric_filtered #sebelum di capping
```

```python
# menghitung dan menampilkan batas atas dan batas bawah untuk mendeteksi outlier dalam data numerik
batas_atas = Q3 + 1.5 * IQR
batas_bawah = Q1 - 1.5 * IQR
batas_data = pd.DataFrame({'Batas Atas': batas_atas, 'Batas Bawah': batas_bawah})
print(batas_data)
```

```python
df_numeric2 = df_numeric.copy()
df_numeric = df.select_dtypes(include=np.number)
df_numeric_filtered = df_numeric[columns_without_opt]
for column in columns_without_opt:
    df_numeric2.loc[df_numeric2[column] > batas_atas[column], column] = batas_atas[column]
    df_numeric2.loc[df_numeric2[column] < batas_bawah[column], column] = batas_bawah[column]
```

```python
# Menggabungkan df_numeric2 yang sudah diperbarui dengan df asli
df[df_numeric2.columns] = df_numeric2
```

```python
# Cek Outlier Tiap Kolom
# Buat figure dan axes (1 baris, 2 kolom)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle("Boxplot Data", fontsize=16)

sns.boxplot(x='Heart Rate', data=df, ax=axes[0]) # Akses subplot pertama
axes[0].set_title('Heart Rate')

plt.show()
```
